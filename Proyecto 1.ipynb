```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
from scipy.sparse import lil_matrix, csr_matrix, diags
from scipy.sparse.linalg import eigsh
from PIL import Image
```

```python
I1 = plt.imread('plane_half.png')
plt.imshow(I1)
plt.show()
I1.shape
```

```python
def color_similarity(color_i, color_j, alpha):
    color_diff = np.linalg.norm(np.array(color_i) - np.array(color_j))
    return alpha * color_diff

def position_similarity(position_i, position_j, beta):
    position_diff = np.linalg.norm(np.array(position_i) - np.array(position_j))
    return beta * position_diff

def similarity(pixel_i, pixel_j, alpha, beta, sigma):
    color_i, position_i = pixel_i
    color_j, position_j = pixel_j

    color_term = color_similarity(color_i, color_j, alpha)
    position_term = position_similarity(position_i, position_j, beta)

    exponent = -1 / (2 * sigma**2)
    return np.exp(exponent * (color_term + position_term))
```

```python
def build_sparse_affinity_matrix(image, alpha, beta, sigma):
    height, width, _ = image.shape
    num_pixels = height * width

    # Inicializa una matriz dispersa con lil_matrix (List of List)
    affinity_matrix = lil_matrix((num_pixels, num_pixels))

    # Función para obtener los vecinos válidos de un píxel
    def get_valid_neighbors(x, y):
        neighbors = []
        if x > 0:
            neighbors.append((x - 1, y))  # Vecino izquierdo
        if x < width - 1:
            neighbors.append((x + 1, y))  # Vecino derecho
        if y > 0:
            neighbors.append((x, y - 1))  # Vecino superior
        if y < height - 1:
            neighbors.append((x, y + 1))  # Vecino inferior
        return neighbors

    # Itera a través de cada píxel
    for y in range(height):
        for x in range(width):
            pixel_i = image[y, x]

            # Calcula la posición del píxel actual en la matriz de afinidad
            pixel_index = y * width + x

            # Calcula la similitud entre el píxel actual y sus vecinos
            for neighbor_x, neighbor_y in get_valid_neighbors(x, y):
                neighbor_pixel = image[neighbor_y, neighbor_x]
                neighbor_index = neighbor_y * width + neighbor_x

                # Calcula la similitud utilizando la función de similitud
                similarity_score = similarity((pixel_i, (x, y)),
                                              (neighbor_pixel, (neighbor_x, neighbor_y)),
                                              alpha, beta, sigma)

                # Actualiza la matriz de afinidad con la similitud calculada
                affinity_matrix[pixel_index, neighbor_index] = similarity_score

    # Convierte la matriz dispersa a un formato más eficiente
    affinity_matrix = affinity_matrix.tocsr()

    return affinity_matrix

# Constantes:
alpha = 1.0
beta = 1.0
sigma = 1.0
image = np.array(I1)
```

```python
# Construye la matriz de afinidad
W = build_sparse_affinity_matrix(image, alpha, beta, sigma)
```

```python
W.shape
```

```python
d = np.array(W.sum(axis=0)).flatten()
```

```python
D = diags(d, 0).tocsr()
```

```python
D[:5,:5].toarray()
```

```python
d_raiz_inversa = 1.0 / np.sqrt(d)
d[:5], d_raiz_inversa[:5]
```

```python
D_raiz_inversa = diags(d_raiz_inversa, 0).tocsr()
```

```python
L = D_raiz_inversa @ (W - D) @ D_raiz_inversa
```

```python
L[:5,:5].toarray()
```

```python
eigenvalores, eigenvectores = eigsh(L, k=2, which='SM')
```

```python
vector_fiedler = eigenvectores[:, 1]
```

```python
mascara_binaria = np.where(vector_fiedler < 0, -1, 1)
```

```python
matriz_binaria = mascara_binaria.reshape(160,240)
```

```python
matrix_blanco_negro = (matriz_binaria + 1) * 127.5
```
